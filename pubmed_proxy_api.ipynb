```python
from fastapi import FastAPI   # Import FastAPI framework
import requests  # Import requests to call PubMed API

app = FastAPI()  # Create a FastAPI app instance (this is our API)

NCBI_API_KEY = "YOUR_API_KEY"  # API key (string) to authenticate requests to PubMed

@app.get("/search_pubmed/")  # Define an endpoint: GET request at "/search_pubmed/"
def search_pubmed(query: str, max_results: int = 10):  # Function takes query (string) and max_results (integer)
    # Construct the PubMed API search URL
    search_url = f"https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&term={query}&retmode=json&retmax={max_results}&api_key={NCBI_API_KEY}"
    
    search_response = requests.get(search_url).json()  # Send request to PubMed API and get JSON response

    pmids = search_response.get("esearchresult", {}).get("idlist", [])  # Extract list of PubMed IDs (PMIDs)
    
    if not pmids:  # If no PMIDs found, return a message
        return {"message": "No results found"}

    pmid_str = ",".join(pmids)  # Convert list of PMIDs to a comma-separated string

    # Construct the PubMed API fetch URL to get article details
    fetch_url = f"https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi?db=pubmed&id={pmid_str}&retmode=json"
    fetch_response = requests.get(fetch_url).json()  # Send request to PubMed API and get JSON response

    results = []  # Create an empty list to store article details

    for pmid in pmids:  # Loop through each PMID
        article = fetch_response["result"].get(pmid, {})  # Get article data from JSON response
        
        results.append({  # Append extracted information to results list
            "title": article.get("title", "No title"),  # Extract title (string)
            "authors": article.get("authors", []),  # Extract authors (list of dictionaries)
            "journal": article.get("source", "Unknown"),  # Extract journal name (string)
            "pub_date": article.get("pubdate", "Unknown"),  # Extract publication date (string)
            "pmid": pmid,  # Store PMID (string)
            "url": f"https://pubmed.ncbi.nlm.nih.gov/{pmid}/"  # Construct article URL (string)
        })

    return {"articles": results}  # Return list of articles in JSON format
```
